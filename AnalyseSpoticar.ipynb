La moyenne des prix des véhicules disponibles sur Spoticar dans un rayon de 150Km autour de Saint-Germain-en-Laye

```python
import seaborn as sns
import matplotlib.pyplot as plt
results["energie"] = results["energie"].replace("Hybride rechargeableHybride rechargeable", "Hybride Rechargeable")
results
```

```python
energyGraph = sns.displot(data=results, x="energie", hue="energie", multiple="stack")
energyGraph.set_xticklabels(rotation=45)
```

Il y a beacoup plus de véhicules 100% thermique en vente sur le marché de l'occasion (Spoticar), que de véhicules à caractère écologique.

```python
graphPrixKm = sns.relplot(x="kilometrage (km)", y="prix (€)", data=results)
plt.xlabel("Kilometrage (Km)", fontsize=16, color="black")
plt.ylabel("Prix (€)", fontsize=16, color="black")
plt.title("Prix / Kilometrage", fontsize=18, color="black")
graphPrixKm.set_xticklabels(rotation=45)
plt.locator_params(nbins=4)
```

Ici, on peut se rendre compte d'une tendance. Les véhicules avec un kilométrage élevé valent en général moins cher que les véhicules avec un kilométrage peu élevé.
